```python
# Python  字符串·列表测试
namesList = ['xiaoWang','xiaoZhang','xiaoHua',12,[255,255,255]]

'''
Python的列表与C语言的数组之间的差别：
    C语言的数组只能存储同一类型的元素；
    Python的列表能存储不同类型的元素；
'''
print("-------------------------------")
print("开始打印列表")
print(namesList[0])
print(namesList[1])

print("-------------------------------")
print("列表的循环测试")
for name in namesList:
    print("for循环输出列表测试：%s"%namesList)
i=0
while i<len(namesList):
    print("while循环输出列表测试：%s"%namesList[i])
    i+=1

print("-------------------------------")
print("给列表添加元素")
temp=input("请输入要添加的学生姓名：")
namesList.append(temp)
print(namesList)

print("-------------------------------")
# 通过extend可以将另一个集合中的元素逐一添加到列表中
print("将一个列表中的元素逐一添加到另一个列表中")
b=["b1","b2","b3"]
namesList.extend(b)
print(namesList)

print("-------------------------------")
print("在列表中查找元素")
# 常用方法有 in,not in, index,count等
findname=input("请输入要找的关键词")
if findname in namesList:
    print("查找关键词成功")
else:
    print("查找关键词失败")

print("-------------------------------")
print("使用index函数查找")
namesList.index("xiaoHua",1,3)

print("-------------------------------")
print("使用del,pop,remove等函数删除元素")
movieName=['加勒比海盗','泰坦尼克号','第一滴血','指环王','速度与激情']
print("----------删除之前------------")
for tempName in movieName:
    print(tempName)

del movieName[1]
print("----------删除之后------------")
for tempName in movieName:
    print(tempName)
```
